```python
#importo le librerie necessarie alla manipolazione dei datasets
import pandas as pd
import numpy as np

#carico il dataset 'dati-classifica-sanremo-1951-2023.xlsx' in un dataframe
df_classifica = pd.read_excel('C:\\Users\\matte\\Downloads\\dati-finalassignment\\dati-classifica-sanremo-1951-2023.xlsx', sheet_name='Sheet1')

#elimino la colonna 'Unnamed: 0'
df_classifica = df_classifica.drop(columns=['Unnamed: 0'])

#rinomino la colonna 'anno' in 'Anno'
df_classifica = df_classifica.rename(columns={'anno':'Anno'})

#rimuovo tutti i duplicati
df_classifica = df_classifica.drop_duplicates()

# rimuovo tutte le righe che hanno valori nulli nella colonna 'Canzone'
df_classifica = df_classifica.dropna(subset=['Canzone'])

#rimuovo tutti gli spazi iniziali e finali di tutte le stringhe
df_classifica = df_classifica.map(lambda x: x.strip() if type(x)==str else x)

#trasformo i valori della colonna 'Anno' in tipo 'data' e la rinomino in 'Data'
df_classifica['Anno'] = pd.to_datetime(df_classifica['Anno'], format='%Y')
df_classifica = df_classifica.rename(columns={'Anno':'Data'})

#in una nuova colonna 'Anno' estraggo l'anno dalla colonna 'Data' come valore intero
df_classifica['Anno'] = df_classifica['Data'].dt.year

#stampo il dataframe
df_classifica
```

```python
# Aggiungo una nuova colonna 'Canzone_Unica' che combina il nome della canzone con l'anno di rilascio, per avere un identificativo unico per ogni canzone
df_classifica['Canzone_Unica'] = df_classifica['Canzone'] + ' (' + df_classifica['Anno'].astype(str) + ')'

#Rimuovi tutti i duplicati dalla colonna 'Canzone_Unica'
df_classifica = df_classifica.drop_duplicates(subset=['Canzone_Unica'])

# Visualizza il dataframe aggiornato
df_classifica
```

```python
#nel df_classifica
#nella colonna autori, sostituisco con una virgola, se presenti, i caratteri che dividono i vari autori
df_classifica['Autori'] = df_classifica['Autori'].str.replace(' ed ', " , ")
df_classifica['Autori'] = df_classifica['Autori'].str.replace(' e ', " , ")
df_classifica['Autori'] = df_classifica['Autori'].str.replace(' & ', " , ")
df_classifica['Autori'] = df_classifica['Autori'].str.replace(' & ', " , ")
df_classifica['Autori'] = df_classifica['Autori'].str.replace(' feat. ', " , ")
df_classifica['Autori'] = df_classifica['Autori'].str.replace(' feat ', " , ")
df_classifica['Autori'] = df_classifica['Autori'].str.replace(' ft. ', " , ")
df_classifica['Autori'] = df_classifica['Autori'].str.replace(' ft ', " , ")
df_classifica['Autori'] = df_classifica['Autori'].str.replace(' con ', " , ")

#creo un nuovo dataframe 'Autori' con la colonna 'Autori' splittata in base alla virgola
df_autori = df_classifica['Autori'].str.split(',', expand=True)
df_autori

#rinomino tutte le colonne presenti nel dataframe 'Autori' aggiungendo prima del numero presente la parola 'Autore'
for i in range(len(df_autori.columns)):
    df_autori = df_autori.rename(columns={i:'Autore_'+str(i+1)})

##aggiungo come prima colonna il nome della canzone
df_autori.insert(0, 'Canzone', df_classifica['Canzone_Unica'])
df_autori

#rimuovo gli spazi iniziali e finali presenti nelle stringhe
df_autori = df_autori.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

#stampo il dataframe
df_autori

```

```python
#nel df_autori voglio visualizzare tutte le righe che hanno valore non nullo nella colonna 'Autore_8'
df_autori[df_autori['Autore_8'].notnull()]
```

```python
#creo  e stampo un dataframe 'autori_canzone' con la colonna 'Autore' che contiene tutti gli autori presenti nel dataframe 'df_autori' accoppiati ad ogni canzone che hanno scritto
autori_canzone = pd.melt(df_autori, id_vars=['Canzone'], value_vars=['Autore_1', 'Autore_2', 'Autore_3', 'Autore_4', 'Autore_5', 'Autore_6', 'Autore_7', 'Autore_8'], value_name='Autore')
autori_canzone = autori_canzone.drop(columns=['variable'])
autori_canzone = autori_canzone.dropna(subset=['Autore'])
autori_canzone = autori_canzone.drop_duplicates()
autori_canzone = autori_canzone.reset_index(drop=True)
autori_canzone
```

```python
#nel df_classifica
#nella colonna Interprete, sostituisco con una virgola, se presenti, i caratteri che dividono i vari interpreti

df_classifica['Interprete'] = df_classifica['Interprete'].str.replace(' ed ', " , ")
df_classifica['Interprete'] = df_classifica['Interprete'].str.replace(' e ', " , ")
df_classifica['Interprete'] = df_classifica['Interprete'].str.replace(' & ', " , ")
df_classifica['Interprete'] = df_classifica['Interprete'].str.replace(' - ', " , ")
df_classifica['Interprete'] = df_classifica['Interprete'].str.replace(' feat. ', " , ")
df_classifica['Interprete'] = df_classifica['Interprete'].str.replace(' feat ', " , ")
df_classifica['Interprete'] = df_classifica['Interprete'].str.replace(' ft. ', " , ")
df_classifica['Interprete'] = df_classifica['Interprete'].str.replace(' ft ', " , ")
df_classifica['Interprete'] = df_classifica['Interprete'].str.replace(' con ', " , ")


#creo un nuovo dataframe 'Interprete' con la colonna 'Interprete' splittata in base alla virgola
df_interpreti = df_classifica['Interprete'].str.split(',', expand=True)

#rinomino tutte le colonne presenti nel dataframe 'Interprete' aggiungendo prima del numero presente la parola 'Interprete'
for i in range(len(df_interpreti.columns)):
    df_interpreti = df_interpreti.rename(columns={i:'Interprete_'+str(i+1)})

#aggiungo come prima colonna il nome della canzone
df_interpreti.insert(0, 'Canzone', df_classifica['Canzone_Unica'])

#rimuovo gli spazi iniziali e finali presenti nelle stringhe
df_interpreti = df_interpreti.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

#stampo il dataframe
df_interpreti
```

```python
#nel df_interprete voglio visualizzare tutte le righe che hanno valore non nullo nella colonna 'Interprete_5'
df_interpreti[df_interpreti['Interprete_5'].notnull()]
```

```python
#creo e stampo un dataframe 'interpreti_canzone' con la colonna 'Interprete' che contiene tutti gli interpreti presenti nel dataframe 'df_interpreti' accoppiati ad ogni canzone che hanno cantato
interpreti_canzone = pd.melt(df_interpreti, id_vars=['Canzone'], value_vars=['Interprete_1', 'Interprete_2', 'Interprete_3', 'Interprete_4', 'Interprete_5'], value_name='Interprete')
interpreti_canzone = interpreti_canzone.drop(columns=['variable'])
interpreti_canzone = interpreti_canzone.dropna(subset=['Interprete'])
interpreti_canzone = interpreti_canzone.drop_duplicates()
interpreti_canzone = interpreti_canzone.reset_index(drop=True)
interpreti_canzone
```

```python
#creo una nuova tabella facendo una inner join tra il dataframe 'df_autori' e il dataframe 'df_interpreti' sulla colonna 'Canzone'
df_autori_interpreti = pd.merge(df_autori, df_interpreti, on='Canzone', how='inner')

#inserisco la colonna 'Anno' dopo la colonna 'Canzone'
df_autori_interpreti.insert(1, 'Anno', df_classifica['Anno'])

#stampo il dataframe
df_autori_interpreti

```

```python
#per ciascun valore del dataframe 'df_autori_interpreti' splitto in base al carattere ' '
df_autori_interpreti_check = df_autori_interpreti.apply(lambda x: x.str.split(' ') if x.dtype == "object" else x)

#definisco una funzione 'check_match' che restituisce True se un autore è anche un interprete, altrimenti restituisce False
def check_match(row):
    autori = [item for sublist in row[['Autore_1', 'Autore_2', 'Autore_3', 'Autore_4', 'Autore_5', 'Autore_6', 'Autore_7', 'Autore_8']].dropna() for item in sublist]
    interpreti = [item for sublist in row[['Interprete_1', 'Interprete_2', 'Interprete_3', 'Interprete_4', 'Interprete_5']].dropna() for item in sublist]
    return any(autore in interpreti for autore in autori)

#applico la funzione 'check_match' al dataframe 'df_autori_interpreti_check' e salvo il risultato nella colonna 'Cantautore'
df_autori_interpreti['Cantautore'] = df_autori_interpreti_check.apply(check_match, axis=1)

#Nella colonna 'Cantautore' sostituisco i valori True con 'Cantautore' e i valori False con 'Cantante'
df_autori_interpreti['Cantautore'] = df_autori_interpreti['Cantautore'].replace({True:'Cantautore', False:'Cantante'})

#stampo il dataframe
df_autori_interpreti
```

```python
#aggiungo al 'df_autori_interpreti' la colonna 'Posizione' presente nel dataframe 'df_classifica' e la inserisco dopo la colonna 'Canzone'
df_autori_interpreti.insert(2, 'Posizione', df_classifica['Posizione'])

#stampo il dataframe
df_autori_interpreti
```
